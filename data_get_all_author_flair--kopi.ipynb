```python
import json
import os
from collections import Counter
from matplotlib import pyplot as plt
```

```python
file_numbers = range(13)
```

```python
import json

save_size = 100000
load_files = './{file_number}.ndjson'
save_folder = 'mbti2022'

# Load ndjson file
authors_flair = {}  # Dictionary to store authors and their flair

for file_number in file_numbers: 
    load_files = f'./{file_number}.ndjson'

    try: 
        with open(load_files, encoding='utf-8') as f:
            for line in f:
                post = json.loads(line)
                author = post['author']

                flair_text = post.get('author_flair_text', None)  # Get the author's flair text, or None if not available

                if flair_text is not None and flair_text != "":
                    authors_flair[author] = flair_text

    except FileNotFoundError:
        print(f"File not found: {load_files}")

# All authors and their flair text
for author, flair in authors_flair.items():
    print(f'Author: {author}, Flair: {flair}') 
```

```python
# Counter to get the number of authors with each unique flair
flair_counts = Counter(authors_flair.values())

# List of flairs and their counts
for flair, count in flair_counts.items():
    print(f'Flair: {flair}, Count: {count}')
```

```python
#Histogram
# Histogram of flair counts
flair_values, flair_counts = zip(*flair_counts.items())  # Unzip the flair counts
plt.bar(flair_values, flair_counts, color='pink', alpha=0.7)
plt.title('Histogram of Personality Types in Reddit Mbti', fontweight = 'bold')
plt.xlabel('Personality type', color='black')
plt.ylabel('Count', color='black')
plt.xticks(rotation=90) 
plt.show()
```

We have made a histogram over the personality types in the Reddit group.
